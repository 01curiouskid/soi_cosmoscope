```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
file = fits.open('bonus_image_1.fits')
image_data = file[0].data

table_data = file[1].data
file.close()
plt.imshow(image_data, cmap='gray')
plt.colorbar()
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Image from Primary HDU')
plt.show()

col1 = table_data['col1']
col2 = table_data['col2']
col3 = table_data['col3']
col4 = table_data['col4']

plt.plot(col1, label='Column 1')
plt.plot(col2, label='Column 2')
plt.plot(col3, label='Column 3')
plt.plot(col4, label='Column 4')

plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()
plt.title('Data from Table HDU')
plt.show()
```
