```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits

file = fits.open('lightcurve_1.fits')
data = file[1].data  # Assuming the light curve data is in the first extension (LIGHTCURVE)
file.info()

# Extract the columns of interest (e.g., 'TIME' and 'SAP_FLUX')
time = data['TIME']
sap_flux = data['SAP_FLUX']

# Create a scatter plot using plt.scatter()
plt.scatter(time, sap_flux)
plt.xlabel('Time')
plt.ylabel('SAP_FLUX')
plt.title('Scatter Plot of TIME vs SAP_FLUX')
plt.show()
```

```python
time = data['TIME']
sap_flux = data['SAP_FLUX']
pdcsap_flux = data['PDCSAP_FLUX']
sap_quality = data['SAP_QUALITY']

# Plot the light curve using plt.plot()
plt.plot(time, sap_flux, label='SAP_FLUX')
plt.plot(time[sap_quality > 0], pdcsap_flux[sap_quality > 0], 'ro', label='PDCSAP_FLUX with SAP_QUALITY > 0')
plt.xlabel('Time')
plt.ylabel('Flux')
plt.legend()
plt.show()
```
